```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm
from scipy import stats

```

```python
df = pd.read_csv('categorical.csv')
```

```python
df
```

```python
def get_bins(n):

    if n > 330:
        return 3
    elif n >32:
        return 2
    elif n > 6:
        return 1
    else:
        return 0 
    
```

```python
df["conflict_bins"]=df["conflict_events"].apply(get_bins)
```

```python
df.head(100)
```

```python
X = df[[
    #'population_density',
     # 'population_growth_annual_per',
   # 'total_population',
     # 'rural_population',
    #'rural_population_per',
   # 'rural_population_growth',
    #'urban_population',
    'corruption_control_estimate',
    'stability_est',
      # 'corruption_control_percentile',
    #'urban_population_per',
   # 'urban_population_growth', 
       #'government_effectiveness_percentile',
    #'stability_percentile',
    #'regulatory_percentile',
   # 'ruleoflaw_percentile',
      # 'regulatory_est',
    #'government _effectiveness_est',  
       # 'ruleoflaw_est',
   # 'accountability_est',
        'ethnic_score',
        'gdp_growth_per',
    #'gdp_percap_per',
      #  'mortality_rate',
    #'gdp_total',
    #'gdp_percap',
    #'gni_total',
      #  'fdi_inflows_gdp',
    #'fdi_inflows_total'
]]

y = df["conflict_bins"]

```

```python
print(df['conflict_bins'].max(),df['conflict_bins'].min() )
print(df['ethnic_score'].max(),df['ethnic_score'].min() )
print(df['mortality_rate'].max(),df['mortality_rate'].min() )
print(df['corruption_control_estimate'].max(),df['corruption_control_estimate'].min() )
print(df['fdi_inflows_gdp'].max(),df['fdi_inflows_gdp'].min() )
print(df['stability_est'].max(),df['stability_est'].min() )
```

```python
from sklearn.model_selection import train_test_split
from tensorflow.keras.utils import to_categorical


X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)

y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

model = Sequential()
model.add(Dense(units=15, activation='relu', input_dim=4))
model.add(Dense(units=15, activation='relu'))
model.add(Dense(units=15, activation='relu'))
model.add(Dense(units=4, activation='softmax'))
```

```python
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])
```

```python
model.fit(
    X_train,
    y_train_categorical,
    epochs=1000,
    shuffle=True,
    verbose=2
)
```

```python
model.summary()
```

```python
import numpy as np
new_data = np.array([[-2.11, -2.04, 0.824, 2.44
 ]])
print(f"Predicted class: {model.predict_classes(new_data)}")

```

```python
model_json= model.to_json()
with open('model.json','w') as json_file:
    json_file.write(model_json)
model.save_weights('model_weights.h5')
```

```python
result = model.predict_classes(new_data)
result_lookup ={
      0:"Stable",
      1:"low risk",
      2:"med risk",
      3:"high risk"}
print(result_lookup[result[0]])
```
