```python
from sqlalchemy import create_engine
import pandas as pd
```

```python
cloud_engine = create_engine(f"postgresql://{remote_db_user}:{remote_db_pwd}@{remote_db_endpoint}:{remote_db_port}/{remote_db_name}")
```

```python
cloud_conn = cloud_engine.connect()
africa_aws = pd.read_sql("SELECT * FROM africa_indicators", cloud_conn)
```

```python
africa_aws
```

```python
import json
import urllib.request
with urllib.request.urlopen('https://raw.githubusercontent.com/codeforamerica/click_that_hood/master/public/data/africa.geojson') as response:
   geojson = json.load(response)
```

```python
africa_aws['Country Name '].unique()
```

```python
names_to_change = ['Congo', 'DR Congo']
new_name = ['Republic of Congo', 'Democratic Republic of Congo']
```

```python
for i in range(len(geojson['features'])):
    

    for j in range(len(names_to_change)):
        if names_to_change[j] == geojson['features'][i]['properties']['name']:
            geojson['features'][i]['properties']['name'] = new_name[j]
```

```python
geojson['features'][53]['properties']
```

```python
africa_GDP_2019 = africa_aws.loc[africa_aws['Year'] == 2019]
africa_GDP_2019 = africa_GDP_2019.loc[africa_GDP_2019['Series Name ']== "GDP per capita,(current US$)"]
africa_GDP_2019.head()
```

```python
for i in range(len(geojson['features'])):
    

    #for j in range(len(africa_GDP_2019['Value'])):
    for index, row in africa_GDP_2019.iterrows():
        if row['Country Name '] == geojson['features'][i]['properties']['name']:
            geojson['features'][i]['properties']['GDP_per_capita_2019'] = row['Value']
```
