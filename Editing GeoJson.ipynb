```python
population = pd.read_sql("select * from population", cloud_conn)
fatalities = pd.read_sql("select * from fatalities", cloud_conn)
conflicts = pd.read_sql("select * from conflict", cloud_conn)
```

```python
population = population[['Country Name',"Country Code",'2018', "Series Name"]]
population = population.loc[population['Series Name'] == 'Population, total']
population = population [["Country Name", "Country Code", "2018"]]
```

```python
fatalities = fatalities[['Country Name','2018']]
conflicts = conflicts[['Country Name', '2018']]
```

```python
choropleth = pd.merge(population, fatalities, on="Country Name")
```

```python
choropleth = choropleth.rename(columns = {"2018_x":"Population_2018","2018_y":"Fatalities_2018"})
```

```python
choropleth = pd.merge(choropleth, conflicts, on="Country Name")
choropleth = choropleth.rename(columns = {"2018":"Conflicts_2018"})
```

```python
population_numbers = []
for item in choropleth['Population_2018']:
    try:
        item = int(item)
        population_numbers.append(item)
    except:
        item = 0
        population_numbers.append(item)
choropleth['Population_2018'] = population_numbers
choropleth['Fatality_Rate_2018'] = choropleth['Fatalities_2018']/choropleth['Population_2018']
```

```python
choropleth = choropleth.rename(columns = {"2018_x":"Conflicts_2018"})
choropleth = choropleth.drop("2018_y", axis=1)
choropleth.head()
```

```python
ethnicity = pd.read_sql("select * from ethnicity", cloud_conn)
ethnicity = ethnicity[["Country Name ", "2018"]]
ethnicity = ethnicity.append({"Country Name ": "South Sudan", "2018":0}, ignore_index=True)
ethnicity = ethnicity.append({"Country Name ": "Equatorial Guinea", "2018":0}, ignore_index=True)
```

```python
ethnicity = ethnicity.rename(columns = {"Country Name ": "Country Name"})
choropleth = pd.merge(choropleth, ethnicity, on="Country Name")
choropleth = choropleth.rename(columns = {"2018":"Homogenous Percentage"})
```

```python
GDP = pd.read_sql("select * from economy", cloud_conn)
GDP.columns
GDP = GDP[["Country Name", "Series Name", "2018.00"]]
```

```python
GDP = GDP.loc[GDP["Series Name"]=="GDP per capita growth (annual %)"]
GDP = GDP.drop("Series Name", axis = 1)
choropleth = pd.merge(choropleth, GDP, on="Country Name")

choropleth = choropleth.drop("2018_y", axis = 1)
#choropleth = choropleth.rename(columns = {"2018_x": "Diversity Percentage"})
choropleth.head()
```

```python
choropleth = choropleth.drop("2018.00_y", axis = 1)
choropleth = choropleth.drop("2018.00_x", axis = 1)
choropleth = choropleth.drop("Series Name_x", axis = 1)
choropleth = choropleth.drop("Series Name_y", axis = 1)
choropleth = choropleth.drop("Series Name", axis = 1)
#choropleth = choropleth.rename(columns = {"2018_x": "Diversity Percentage"})
choropleth.head()
```

```python
choropleth = pd.merge(choropleth, GDP, on="Country Name")
choropleth.head()
```

```python
for i in range(len(geojson['features'])):
    

    for index, row in choropleth.iterrows():
        if row['Country Name'] == geojson['features'][i]['properties']['name']:
            geojson['features'][i]['properties']['Conflicts_2018'] = row['Conflicts_2018']
            geojson['features'][i]['properties']['Fatalities_2018'] = row['Fatalities_2018']
            geojson['features'][i]['properties']['Fatalities_Rate_2018'] = row['Fatality_Rate_2018']
            geojson['features'][i]['properties']['Diversity_Percentage'] = row['Diversity Percentage']
            geojson['features'][i]['properties']['GDP_Growth_2018'] = row['2018.00']
geojson['features'][1]['properties']
```

```python
with open('AfricaGeoJson.json', 'w') as json_file:
    json.dump(geojson, json_file)
```
